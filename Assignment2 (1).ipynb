# **1.List Creation:**
---



```python
age_li=[28,29,26,27,28]
name_li=["divya","gowshi","arun","balu","Mano"]
print(age_li)
```

```python
#append the string
name_li.append("Yazhini")
name_li
```

```python
# Insert the element 30 at index 2 in age_list
age_li.insert(2,30)
age_li
```

```python

# Remove the string "Yazhini" from name_list
name_li.remove("Yazhini")
name_li
```

```python
#Pop the last element from age_list
removed_age=age_li.pop(5)
print(removed_age)
```

```python
#Extend the age_list with additional ages [29, 30, 26]
age_li.extend([29, 30, 26])
print(age_li)
```

```python
#Sort age_list in descending order
age_li.sort(reverse=True)
age_li
```

```python
#Find Max age, Min age and sum of all ages from age_list

print(f"max_age:{max(age_li)}")
print(f"min_age:{min(age_li)}")
print(f"sum of all ages:{sum(age_li)}")
```

# 3. Accessing List Elements:

```python
#Print the first element of name_list
print(f"first element of name_list : {name_li[0]}")
#Print the last element of name_list
print(f"Last element of name_list : {name_li[-1]}")
#Print the elements from index 2 to index 4 in name_list
print(f"elements from index 2 to index 4 in name_list : {name_li[1:5]}")
#. Print the elements of name_list in reverse order
print(f"elements of name_list in reverse order : {name_li[::-1]}")
```



# Dictionary (Creation, Modification and Access):



```python
#Create a dictionary named student_marks that maps the names of five students to their marks (use scale of from 0 to 100)

student_marks={"Dev":85,
               "Dee":99,
               "Mani":100,
               "Raju":98,
               "Diya":95}
print(student_marks)
```

```python
#Access and print the mark of a specific student, of your choice
print(f"Dee marks: {student_marks['Dee']}")
```

```python
#Add a new student "Janani" with a mark of 80 to the student_marks dictionary
student_marks["Janani"]=80
print(student_marks)
```

```python
#Update the mark of any one older student to 82
student_marks["Mani"]=82
student_marks
```

```python
#Use the keys(), values(), and items() methods to print all keys, values, and key-value pairs in the student_marks dictionary
student_marks.keys()
student_marks.values()
student_marks.items()
```

# Sets (Operations):  

```python
#Create a set called my_set with following values

my_li={"a","e","i","o","u","a","a","i"}
my_set=set(my_li)
print(my_set)
#elplanation:
#1.Sets only store unique elements
#2.All duplicates are removed
#3.Sets are unordered
```

```python
#Attempt to change the value of my_set[4] = 's'. If code throws an error, provide an explanation

my_set[4]="s"
#elplanation:
#Python sets are unordered collections and do not support indexing, so item assignment like my_set[4] = 's' is not allowed
```

```python
set1={1, 3, 5, 7, 9}
set2={2, 3, 5, 8, 10}
set1
set2
```

```python
#Compute and print the union and intersection of set1 and set2


set1.union(set2)
set1.intersection(set2)
```
